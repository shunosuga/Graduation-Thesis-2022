```julia
dir = "/home/shunosuga/data/sotsuron/kernel_conv1_sort"

cade = os.listdir(dir)

println(cade)
```

```julia
mean_caffe_raw_s = np.load(os.path.join(dir, cade[1]))
ind_caffe = np.load(os.path.join(dir, cade[2]))
mean_caffe_abs_s = np.load(os.path.join(dir, cade[3]))
mean_unet_raw_s = np.load(os.path.join(dir, cade[4]))
ind_unet = np.load(os.path.join(dir, cade[5]))
mean_unet_abs_s = np.load(os.path.join(dir, cade[6]))
```

```julia
x1 = range(1, 96, step=1)
y11 = mean_caffe_raw_s[1, :]
y12 = mean_caffe_abs_s[3, :]

plot(
    x1, y11, 
    
    ylims=(-0.025, 0.52), 
    title="CaffeNet",
    titlefont=font(12, "helvetica"),
    ylabel="Profile correlation",
    yguidefontsize=10,
    xticks = (
        Vector(1:96), 
        ["" for n=1:96]
    ),
    grid=(:y),
    label="Raw", 
    color=:blue,
    legend= :topleft,
    size=(420, 280),
    fontfamily="helvetica",
    linewidth=1.2,
    background_color_legend = nothing,
    foreground_color_legend = nothing
    )

plot!(
    x1, y12, 
    label="Absolute",
    color=palette(:lightrainbow),
    linewidth=1.2
    )
```

```julia
x2 = range(1, 96, step=1)
y11 = mean_caffe_raw_s[4, :]
y12 = mean_caffe_abs_s[6, :]

plot(
    x2, y11, 
    ylims=(-0.025, 0.52), 
    title="CaffeU-Net",
    titlefont=font(12, "helvetica"),
    ylabel="Profile correlation",
    yguidefontsize=10,
    xticks = (
        Vector(1:96), 
        ["" for n=1:96]
    ),
    grid=(:y),
    color=:blue,
    legend= false,
    size=(420, 280),
    fontfamily="helvetica",
    linewidth=1.2
    )
    
plot!(
    x2, y12, 
    color=palette(:lightrainbow),
    linewidth=1.2
    )
```
